```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


tips = sns.load_dataset("tips")

```

```python
tips.head()
```

```python
sns.catplot(data=tips, x="day", y="total_bill",hue=tips['sex'])
```

```python
sns.catplot(data=tips, x="day", y="total_bill", jitter=False,hue=tips['sex'])
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box" ,hue=tips['smoker'])
```
