```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Sample data
data = {
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Female'],
    'Age': [25, 30, 35, 40, 28],
    'Purchased': [0, 1, 0, 1, 1]  # Target variable
}

df = pd.DataFrame(data)


```

```python
df.head()
```

```python
X = df[['City', 'Gender', 'Age']]  # Features
y = df['Purchased']  # Target variable
```

```python

encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(X[['City', 'Gender']]).toarray()

# Combine one-hot encoded columns with numerical features
feature_names = encoder.get_feature_names_out(['City', 'Gender'])
encoded_df = pd.DataFrame(encoded_features, columns=feature_names)

# Add the numerical column
X_encoded = pd.concat([encoded_df, X[['Age']].reset_index(drop=True)], axis=1)
```

```python
X_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```
