```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4]
y=[1,12,13,14]
plt.title('Line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[1,12,13,14]
plt.title('Line plot',loc='right')# left,center
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
plt.show()
```
